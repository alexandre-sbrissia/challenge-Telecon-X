```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# URLs das lojas
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Carregar os dados
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

print("‚úÖ Dados carregados!")
print(f"Loja 1: {len(loja1)} vendas")
print(f"Loja 2: {len(loja2)} vendas")
print(f"Loja 3: {len(loja3)} vendas")
print(f"Loja 4: {len(loja4)} vendas")
```

```python
# AN√ÅLISE 1: FATURAMENTO TOTAL
print("="*50)
print("üí∞ FATURAMENTO TOTAL POR LOJA")
print("="*50)

fat1 = loja1['Pre√ßo'].sum()
fat2 = loja2['Pre√ßo'].sum()
fat3 = loja3['Pre√ßo'].sum()
fat4 = loja4['Pre√ßo'].sum()

print(f"Loja 1: R$ {fat1:,.2f}")
print(f"Loja 2: R$ {fat2:,.2f}")
print(f"Loja 3: R$ {fat3:,.2f}")
print(f"Loja 4: R$ {fat4:,.2f}")
```

```python
# AN√ÅLISE 2: CATEGORIAS MAIS POPULARES
print("="*50)
print("üì¶ TOP 3 CATEGORIAS POR LOJA")
print("="*50)

print("\nLOJA 1:")
print(loja1['Categoria do Produto'].value_counts().head(3))

print("\nLOJA 2:")
print(loja2['Categoria do Produto'].value_counts().head(3))

print("\nLOJA 3:")
print(loja3['Categoria do Produto'].value_counts().head(3))

print("\nLOJA 4:")
print(loja4['Categoria do Produto'].value_counts().head(3))
```

```python
# AN√ÅLISE 3: M√âDIA DE AVALIA√á√ÉO
print("="*50)
print("‚≠ê M√âDIA DE AVALIA√á√ÉO")
print("="*50)

av1 = loja1['Avalia√ß√£o da compra'].mean()
av2 = loja2['Avalia√ß√£o da compra'].mean()
av3 = loja3['Avalia√ß√£o da compra'].mean()
av4 = loja4['Avalia√ß√£o da compra'].mean()

print(f"Loja 1: {av1:.2f}")
print(f"Loja 2: {av2:.2f}")
print(f"Loja 3: {av3:.2f}")
print(f"Loja 4: {av4:.2f}")
```

```python
# AN√ÅLISE 4: PRODUTO MAIS VENDIDO
print("="*50)
print("üèÜ PRODUTO MAIS VENDIDO POR LOJA")
print("="*50)

print("\nLOJA 1:", loja1['Produto'].value_counts().index[0])
print("LOJA 2:", loja2['Produto'].value_counts().index[0])
print("LOJA 3:", loja3['Produto'].value_counts().index[0])
print("LOJA 4:", loja4['Produto'].value_counts().index[0])
```

```python
# AN√ÅLISE 5: CUSTO M√âDIO DO FRETE
print("="*50)
print("üöö CUSTO M√âDIO DO FRETE")
print("="*50)

frete1 = loja1['Frete'].mean()
frete2 = loja2['Frete'].mean()
frete3 = loja3['Frete'].mean()
frete4 = loja4['Frete'].mean()

print(f"Loja 1: R$ {frete1:.2f}")
print(f"Loja 2: R$ {frete2:.2f}")
print(f"Loja 3: R$ {frete3:.2f}")
print(f"Loja 4: R$ {frete4:.2f}")
```

```python
# TABELA RESUMO
print("="*70)
print("üìä RESUMO COMPARATIVO DAS LOJAS")
print("="*70)

print(f"{'Loja':<6} {'Faturamento':>15} {'Avalia√ß√£o':>12} {'Frete M√©dio':>15} {'Top Categoria':>20}")
print("-"*70)

print(f"{'Loja 1':<6} {fat1:>15,.2f} {av1:>12.2f} {frete1:>15.2f} {loja1['Categoria do Produto'].value_counts().index[0]:>20}")
print(f"{'Loja 2':<6} {fat2:>15,.2f} {av2:>12.2f} {frete2:>15.2f} {loja2['Categoria do Produto'].value_counts().index[0]:>20}")
print(f"{'Loja 3':<6} {fat3:>15,.2f} {av3:>12.2f} {frete3:>15.2f} {loja3['Categoria do Produto'].value_counts().index[0]:>20}")
print(f"{'Loja 4':<6} {fat4:>15,.2f} {av4:>12.2f} {frete4:>15.2f} {loja4['Categoria do Produto'].value_counts().index[0]:>20}")
```

```python
# GR√ÅFICO 1: FATURAMENTO
plt.figure(figsize=(10,5))

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [fat1, fat2, fat3, fat4]
cores = ['green', 'blue', 'orange', 'red']

barras = plt.bar(lojas, faturamentos, color=cores)
plt.title('Faturamento Total por Loja', fontsize=16)
plt.xlabel('Loja')
plt.ylabel('Faturamento (R$)')

# Colocar valores nas barras
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2., altura + 500,
             f'R${altura:,.0f}', ha='center', va='bottom')

plt.show()
```

```python
# GR√ÅFICO 2: CATEGORIAS DA LOJA COM MENOR FATURAMENTO

# Descobrir qual loja tem menor faturamento
fats = [fat1, fat2, fat3, fat4]
menor_fat = min(fats)

if menor_fat == fat1:
    loja_escolhida = loja1
    nome_loja = "Loja 1"
elif menor_fat == fat2:
    loja_escolhida = loja2
    nome_loja = "Loja 2"
elif menor_fat == fat3:
    loja_escolhida = loja3
    nome_loja = "Loja 3"
else:
    loja_escolhida = loja4
    nome_loja = "Loja 4"

print(f"üìå A loja com menor faturamento √©: {nome_loja}")

# Gr√°fico de pizza
categorias = loja_escolhida['Categoria do Produto'].value_counts()
top5 = categorias.head(5)

plt.figure(figsize=(8,8))
plt.pie(top5.values, labels=top5.index, autopct='%1.1f%%', startangle=90)
plt.title(f'Categorias mais vendidas - {nome_loja}', fontsize=14)
plt.show()
```

```python
# GR√ÅFICO 3: FRETE M√âDIO
plt.figure(figsize=(10,5))

fretes = [frete1, frete2, frete3, frete4]
barras = plt.bar(lojas, fretes, color=cores)
plt.title('Custo M√©dio do Frete por Loja', fontsize=16)
plt.xlabel('Loja')
plt.ylabel('Frete M√©dio (R$)')

for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2., altura + 0.5,
             f'R${altura:.2f}', ha='center', va='bottom')

plt.show()
```

```python
# AN√ÅLISE FINAL - QUAL LOJA VENDER?

print("="*70)
print("üîç DECIS√ÉO FINAL - QUAL LOJA VENDER?")
print("="*70)

# Normalizar as notas (0 a 1)
max_fat = max(fats)
nota_fat1 = fat1 / max_fat
nota_fat2 = fat2 / max_fat
nota_fat3 = fat3 / max_fat
nota_fat4 = fat4 / max_fat

# Avalia√ß√£o (maior √© melhor)
max_av = max([av1, av2, av3, av4])
nota_av1 = av1 / max_av
nota_av2 = av2 / max_av
nota_av3 = av3 / max_av
nota_av4 = av4 / max_av

# Frete (menor √© melhor)
min_frete = min([frete1, frete2, frete3, frete4])
nota_frete1 = min_frete / frete1
nota_frete2 = min_frete / frete2
nota_frete3 = min_frete / frete3
nota_frete4 = min_frete / frete4

# Nota final (m√©dia das 3 notas) - QUANTO MENOR, PIOR A LOJA
nota_final1 = (nota_fat1 + nota_av1 + nota_frete1) / 3
nota_final2 = (nota_fat2 + nota_av2 + nota_frete2) / 3
nota_final3 = (nota_fat3 + nota_av3 + nota_frete3) / 3
nota_final4 = (nota_fat4 + nota_av4 + nota_frete4) / 3

# Mostrar notas
print(f"\nüìä NOTAS DE DESEMPENHO (0 a 1):")
print(f"Loja 1: {nota_final1:.3f}")
print(f"Loja 2: {nota_final2:.3f}")
print(f"Loja 3: {nota_final3:.3f}")
print(f"Loja 4: {nota_final4:.3f}")

# Descobrir a pior loja (menor nota)
notas = [nota_final1, nota_final2, nota_final3, nota_final4]
pior_nota = min(notas)

if pior_nota == nota_final1:
    pior_loja = "Loja 1"
elif pior_nota == nota_final2:
    pior_loja = "Loja 2"
elif pior_nota == nota_final3:
    pior_loja = "Loja 3"
else:
    pior_loja = "Loja 4"

print(f"\n‚ùå A LOJA COM PIOR DESEMPENHO √â: {pior_loja}")
```
